# Attack MSU trained model

```python
import os
import logging

import shutil
import warnings
import splitfolders

warnings.filterwarnings('ignore')

import tensorflow as tf
import numpy as np
from matplotlib import pyplot as plt

from art.attacks.evasion import FastGradientMethod, CarliniLInfMethod
```

```python
if tf.__version__[0] != '2':
    raise ImportError('This notebook requires TensorFlow v2.')
```

# Load dataset

```python
base_dir = './dataset'
data_dir = './dataset/data'
```

```python
train_dir = os.path.join(data_dir, 'train') ## Train Dataset
validation_dir = os.path.join(data_dir, 'val') ## Validation Dataset
test_dir = os.path.join(data_dir, 'test') ## Test Dataset
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale = 1./255,
    rotation_range = 20,
    width_shift_range = 0.1,
    height_shift_range = 0.2,
    horizontal_flip = True,
    shear_range = 0.2,
    zoom_range = 0.2,
    fill_mode = 'nearest'
)

validation_datagen = ImageDataGenerator(
    rescale = 1./255,
)

test_datagen = ImageDataGenerator(
    rescale = 1./255,
)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size = (150, 150),
    batch_size = 32,
    class_mode = 'categorical'
)

validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size = (150, 150),
    batch_size = 32,
    class_mode = 'categorical'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size = (150, 150),
    batch_size = 100,
    class_mode = 'categorical'
)
```

```python
train_generator.class_indices
```

Evaluate model accuracy on test data.

```python
model = tf.keras.models.load_model('vgg19_msu_categorical.keras')
```

```python
# evaluate the network
scores = model.evaluate(validation_generator)
print("%s%s: %.2f%%" % ("Evaluate ", model.metrics_names[1], scores[1]*100))
```

Create a ART Keras classifier for the TensorFlow Keras model.

```python
from art.estimators.classification import TensorFlowV2Classifier
from tensorflow.keras.losses import CategoricalCrossentropy

loss_object = CategoricalCrossentropy()

classifier = TensorFlowV2Classifier(model=model,
                                    clip_values=(0, 1), 
                                    nb_classes=2, 
                                    input_shape=(150, 150, 3), 
                                    loss_object=loss_object)
```

```python
def plot_adversarial_examples(original_images, adv_images, original_labels, adv_labels, class_names, num_examples=3):
    # Plot a grid of 2x3 (3 original images, 3 adversarial images)
    plt.figure(figsize=(10, 7))

    for i in range(num_examples):
        # Plot original image
        plt.subplot(2, num_examples, i + 1)
        plt.imshow(original_images[i].astype("float32"))
        plt.title(f"Original: {class_names[np.argmax(original_labels[i])]}")
        plt.axis('off')
        
        # Plot adversarial image
        plt.subplot(2, num_examples, i + 1 + num_examples)
        plt.imshow(adv_images[i].astype("float32"))
        plt.title(f"Adversarial: {class_names[np.argmax(adv_labels[i])]}")
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python
class_names = list(train_generator.class_indices.keys())
class_names
```

# Get samples

```python
x_test, y_test = next(test_generator)
```

```python
np.savez('attacks/ori.npz', x_test, y_test)
```

```python
data = np.load('attacks/ori.npz')
```

```python
x_test, y_test = data['arr_0'], data['arr_1']
```

# Fast Gradient Sign Method attack

Create a ART Fast Gradient Sign Method attack.

```python
attack_fgsm = FastGradientMethod(estimator=classifier, eps=0.1)
```

Generate adversarial test data.

```python
x_test_adv = attack_fgsm.generate(x_test)
```

```python
np.save('attacks/fgsm.npy', x_test_adv)
```

```python
y_test_adv = classifier.predict(x_test_adv)
```

Evaluate accuracy on adversarial test data and calculate average perturbation.

```python
loss_test, accuracy_test = model.evaluate(x_test_adv, y_test)
perturbation = np.mean(np.abs((x_test_adv - x_test)))
print('Accuracy on adversarial test data: {:4.2f}%'.format(accuracy_test * 100))
print('Average perturbation: {:4.2f}'.format(perturbation))
```

The result of 18.00% accuracy on adversarial test data with an average perturbation of 0.09 suggests that the model's performance is significantly compromised when subjected to adversarial attacks generated using the Fast Gradient Sign Method (FGSM). The low accuracy indicates that the model struggles to correctly classify the adversarial examples, which are crafted by making small, targeted changes to the original input data. The average perturbation of 0.09 reflects the magnitude of these changes, implying that only slight modifications were needed to fool the model effectively.

This demonstrates that the model is vulnerable to FGSM attacks, even with relatively small perturbations, emphasizing the need for enhanced robustness against adversarial inputs.

Visualise adversarial test samples.

```python
plot_adversarial_examples(x_test, x_test_adv, y_test, y_test_adv, class_names, num_examples=3)
```

# Carlini&Wagner Linf

Create a ART Carlini&Wagner Infinity-norm attack.

```python
attack_cw = CarliniLInfMethod(classifier=classifier,
                              max_iter=100,
                              learning_rate=0.01,
                              initial_const=1e0,
                              largest_const=2e0)
```

Generate adversarial test data.

```python
x_test_cw = attack_cw.generate(x_test)
```

```python
np.save('attacks/cwlinf.npy', x_test_cw)
```

```python
y_test_cw = classifier.predict(x_test_cw)
```

Evaluate accuracy on adversarial test data and calculate average perturbation.

```python
loss_test, accuracy_test = model.evaluate(x_test_cw, y_test)
perturbation = np.mean(np.abs((x_test_cw - x_test)))
print('Accuracy on adversarial test data: {:4.2f}%'.format(accuracy_test * 100))
print('Average perturbation: {:4.2f}'.format(perturbation))
```

The result of 0.00% accuracy on adversarial test data with an average perturbation of 0.01 using the Carlini-Wagner (CW) ℓ∞ attack indicates that the model is completely vulnerable to this type of adversarial attack. The 0% accuracy means that the model fails to correctly classify any of the adversarial examples.

Despite the small perturbation (average 0.01), the CW attack is highly effective, showing that even minimal modifications to the input data are sufficient to completely mislead the model. This highlights the potency of the CW attack and reveals a critical weakness in the model's robustness against adversarial manipulation.

Visualise adversarial test samples.

```python
plot_adversarial_examples(x_test, x_test_cw, y_test, y_test_cw, class_names, num_examples=3)
```

# Elastic Net Attack L1

```python
from art.attacks.evasion import ElasticNet
```

```python
attack_ead = ElasticNet(
                    classifier=classifier,
                    targeted=False,  # Use True for targeted attacks
                    max_iter=100,
                    beta=0.1,  # L1 regularization
                    decision_rule='L1',  # L1 decision rule
                    verbose=True
                    )
```
